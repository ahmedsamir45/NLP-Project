# Edit the text

```python
import PyPDF2 as pdf
file = open("Definitions about Data Science.pdf","rb")
```

```python
reader=pdf.PdfReader(file)
reader.pages[0].extract_text()
```

```python
string=""

for i in range(len(reader.pages)):
    string+=reader.pages[i].extract_text()
    
```

```python
string = string.replace("\n"," ").replace("•"," ")
```

```python
string
```

# tokenization & pos 

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp(string)
type(doc)
```

```python
index=0
for token in doc:
    print(index)
    print(token.text)
    print(token.shape)
    print(token.shape_)
    print(token.is_alpha)
    print(token.is_stop)
    index+=1
    print("="*50)
    print("")
    
```

## token with sentence

```python
for sent in doc.sents:
    print(sent,"\n")
    
```

```python
doc_sents = [sent for sent in doc.sents]
doc_sents
```

## pos

```python
for token in doc[:5]:
    print("word: ",token.text, "type: ",token.pos_,"tag: ",token.tag_,"means: ",spacy.explain(token.tag_))
    print("-"*50)
    print( )
```

```python
pos_counts = doc.count_by(spacy.attrs.DEP)
for k,v in sorted(pos_counts.items()):
    print(f"{k} {doc.vocab[k].text:{10}}:{v}")
```

```python
import nltk
from nltk.tokenize import PunktSentenceTokenizer

cusom_sent_tokenizer= PunktSentenceTokenizer(string)
tokenized = cusom_sent_tokenizer.tokenize(string)[:1]

for i in tokenized:
    for w,m in nltk.pos_tag(nltk.word_tokenize(i)):
        print(f"{w} \t {m}\t {spacy.explain(m)}")
        print("")
        print("="*50)
```

# Stemming & Lemmatization

```python
# doc is the string after tokenize
from nltk.stem import PorterStemmer, SnowballStemmer, LancasterStemmer
from nltk.tokenize import sent_tokenize, word_tokenize

p_s = PorterStemmer()
s_stem = SnowballStemmer(language = 'english')
l_s = LancasterStemmer()
```

```python
print("{0:20}{1:20}{2:20}{3:20}".format("Word","Porter Stemmer","SnowBall Stemmer","lancaster Stemmer"))

for word in word_tokenize(string)[:10]:
    print("{0:20}{1:20}{2:20}{3:20}".format(word,p_s.stem(word),s_stem.stem(word),l_s.stem(word)))
```

```python
for token in doc[:5]:
    print(token.text, '\t', token.pos_, '\t', token.lemma, '\t', token.lemma_)
    print(" ")
```

```python
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
punctuations="?:!.,;"

sentence_words=word_tokenize(string)
for word in sentence_words:
    if word in punctuations:
        sentence_words.remove(word)
```

```python
print("{0:20}{1:20}".format("Word","Lemma"))
for word in sentence_words[:10]:
    print ("{0:20}{1:20}".format(word,wordnet_lemmatizer.lemmatize(word, pos="v")))
```
